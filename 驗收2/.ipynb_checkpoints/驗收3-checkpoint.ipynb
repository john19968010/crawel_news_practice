```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
from selenium.webdriver.firefox.options import Options 
```

```python
# https://gogakuru.com/english/phrase/genre/180_%E5%88%9D%E7%B4%9A%E3%83%AC%E3%83%99%E3%83%AB.html?pageID=1&flow=enSearchGenre&condGenre=180&layoutPhrase=1&orderPhrase=1&condMovie=0&perPage=50
#抓取資料

#讓瀏覽器在背景執行（宣告options）
options = webdriver.ChromeOptions() 
options.add_argument("--headless")
chrome_driver_path = "/Users/john19968010/Desktop/chromedriver"
# 將options加入Chrome裡,driver用executable_path宣告進入
driver = webdriver.Chrome(chrome_options = options, executable_path = chrome_driver_path) 
driver.implicitly_wait(10)
result = []
for i in range(1,188): #因總共有187頁,網野名稱只有id的數字不一樣 
    url = f'https://gogakuru.com/english/phrase/genre/180_%E5%88%9D%E7%B4%9A%E3%83%AC%E3%83%99%E3%83%AB.html?pageID={i}&flow=enSearchGenre&condGenre=180&layoutPhrase=1&orderPhrase=1&condMovie=0&perPage=50'
    driver.get(url)
    #休息5秒讓網頁跑完
    time.sleep(5)
    items = driver.find_elements_by_xpath("//span[@class='font-en']")
    for item in items:
        result.append(item.text)
driver.quit()


```

```python
#轉成txt檔
with open('EXAM3.txt','a') as file:
    for i in result:
        file.write(i+'\n')
```

```python
#https://docs.python.org/3/library/pickle.html?highlight=pickle#pickle.Pickler
#轉成pickle檔 (參照官方文件裡的EXAMPLE)
#寫入
import pickle
with open('EXAM3.pickle','wb') as file:
    pickle.dump(result,file, pickle.HIGHEST_PROTOCOL)
```

```python
#讀出
import pickle
with open('EXAM3.pickle','rb') as file:
    data = pickle.load(file,encoding = 'utf8')
# data
```

```python
#利用py在mysql創建table
try:
    conn = mysql.connector.connect(host="192.168.56.31", user="john", passwd="john", db="test", charset="utf8")
    cursor = conn.cursor()  #準備連結 
    sql = "create table Word(id int AUTO_INCREMENT primary key,word Varchar(40)); "
   
    
    cursor.execute(sql)  # 匯入SQL指令
    conn.commit()
    print("新增成功")
except mysql.connector.Error as e:
    print(str(e))
    print("新增失敗")
except Exception as e:
    print(str(e))
    conn.rollback()
finally:
    if conn is not None:
        conn.close()
```

```python
#利用py在mysql匯入資料
try:
    conn = mysql.connector.connect(host="192.168.56.31", user="john", passwd="john", db="test", charset="utf8")
    cursor = conn.cursor()  
    for i in data:
        sql = '''insert into Word(word) values ("'''+i+'''")'''
        #要把匯入資料的執行指令寫在for迴圈裡,不然只會有最後一筆
        cursor.execute(sql)  
        conn.commit()
    print("新增成功")
except mysql.connector.Error as e:
    print(str(e))
    print("新增失敗")
except Exception as e:
    print(str(e))
    conn.rollback()
finally:
    if conn is not None:
        conn.close()
```

```python
#因部分內容裡包含雙引號,這種符號不是str無法被寫入sql
#所以用replace語法將其刪除
dataN = []
for i in data:
    a = i.replace('"','')
    b = a.replace('\\','')
    dataN.append(b)
# print(dataN)
```

```python
#再次將資料匯入mysql
try:
    conn = mysql.connector.connect(host="192.168.56.31", user="john", passwd="john", db="test", charset="utf8")
    cursor = conn.cursor()  
    #匯入的東西變dataN
    for i in dataN:
        
        sql = '''insert into Word(word) values ("'''+i+'''")'''
        cursor.execute(sql)  
        conn.commit()
    print("新增成功")
except mysql.connector.Error as e:
    print(str(e))
    print("新增失敗")
except Exception as e:
    print(str(e))
    conn.rollback()
finally:
    if conn is not None:
        conn.close()
```
